```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
data_new_short_genomic = pd.read_csv('short-genomic.csv')

# Split the dataset into 'real' and 'simulated' based on the 'Type' column
simulated_data = data_new_short_genomic[data_new_short_genomic['Type'] == 'simulated']
real_data = data_new_short_genomic[data_new_short_genomic['Type'] == 'real']

# Clean the data by ensuring valid string values and converting 'Accuracy' to numeric
real_data_cleaned = real_data.dropna(subset=['Unnamed: 0', 'Accuracy'])
real_data_cleaned['Accuracy'] = pd.to_numeric(real_data_cleaned['Accuracy'], errors='coerce')
real_data_cleaned = real_data_cleaned.dropna(subset=['Accuracy'])

# Plot with x-axis ticks labeled as 0.5, 1.0, 1.5, etc., and show 'x 1e7' under x-axis
plt.figure(figsize=(8, 6))
bars_real = plt.barh(real_data_cleaned['Unnamed: 0'], real_data_cleaned['Accuracy'], color=['red', 'black', 'orange', 'green', 'blue'])
plt.xlabel('Alignment Matches')
plt.title('Alignment Matches')

# Set the x-axis ticks as 0.5, 1.0, 1.5, etc.
plt.xticks(ticks=[0.5e7, 1e7, 1.5e7, 2e7, 2.5e7, 3e7], labels=[0.5, 1.0, 1.5, 2.0, 2.5, 3.0])

# Add label (x 1e7) below the x-axis
plt.figtext(1, 0.05, 'x 1e7', fontsize=10, ha='right')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Load the dataset (assuming it's already loaded as 'data_new')
data_new = pd.read_csv('short-genomic.csv')

# Clean the data and convert the 'Time(s)' column to numeric, setting non-numeric values as NaN
data_new['Time(s)'] = pd.to_numeric(data_new['Time(s)'], errors='coerce')

# Separate the data into simulated and real datasets
simulated_data_cleaned = data_new[data_new['Type'] == 'simulated']
real_data_cleaned = data_new[data_new['Type'] == 'real']

# Convert time to minutes
simulated_data_cleaned['Time(min)'] = simulated_data_cleaned['Time(s)'] / 60
real_data_cleaned['Time(min)'] = real_data_cleaned['Time(s)'] / 60

# Plot for simulated tools with time values
plt.figure(figsize=(8, 6))
bars_simulated = plt.bar(simulated_data_cleaned['Unnamed: 0'], simulated_data_cleaned['Time(min)'], 
                         color=['red', 'blue', 'green', 'orange', 'purple', 'brown'], yerr=simulated_data_cleaned['Time(min)']*0.1)
plt.xlabel('Tools')
plt.ylabel('Run time (min)')
plt.title('Run time for Simulated Tools')

# Display time values on each bar
for bar in bars_simulated:
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 5, f'{bar.get_height():.0f}', ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.show()

# Plot for real tools with time values
plt.figure(figsize=(8, 6))
bars_real = plt.bar(real_data_cleaned['Unnamed: 0'], real_data_cleaned['Time(min)'], 
                    color=['red', 'blue', 'green', 'orange', 'purple', 'brown'], yerr=real_data_cleaned['Time(min)']*0.1)
plt.xlabel('Tools')
plt.ylabel('Run time (min)')
plt.title('Run time for Real Tools')

# Display time values on each bar
for bar in bars_real:
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 5, f'{bar.get_height():.0f}', ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.show()
```
