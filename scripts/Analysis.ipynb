```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Load the CSV files
file_path_minimap2 = 'minimap2.csv'
file_path_vat = 'vat.csv'

df_minimap2 = pd.read_csv(file_path_minimap2)
df_vat = pd.read_csv(file_path_vat)

# Extracting and cleaning data for Minimap2 Marine
minimap2_marine = df_minimap2.iloc[1:, 0:4]
minimap2_marine.columns = ['window_size', 'Time(s)', 'Accuracy(%)', 'kmer']
minimap2_marine['Dataset'] = 'Marine'
minimap2_marine['Tool'] = 'Minimap2'

# Extracting and cleaning data for Minimap2 Cami
minimap2_cami = df_minimap2.iloc[1:, 5:9]
minimap2_cami.columns = ['window_size', 'Time(s)', 'Accuracy(%)', 'kmer']
minimap2_cami['Dataset'] = 'Cami'
minimap2_cami['Tool'] = 'Minimap2'

# Extracting and cleaning data for VAT Marine
vat_marine = df_vat.iloc[1:, 0:4]
vat_marine.columns = ['window_size', 'Time(s)', 'Accuracy(%)', 'kmer']
vat_marine['Dataset'] = 'Marine'
vat_marine['Tool'] = 'VAT'

# Extracting and cleaning data for VAT Cami
vat_cami = df_vat.iloc[1:, 5:9]
vat_cami.columns = ['window_size', 'Time(s)', 'Accuracy(%)', 'kmer']
vat_cami['Dataset'] = 'Cami'
vat_cami['Tool'] = 'VAT'

# Combining all data into a single dataframe
combined_df = pd.concat([minimap2_marine, minimap2_cami, vat_marine, vat_cami], ignore_index=True)
combined_df = combined_df.dropna().astype({'window_size': 'int', 'Time(s)': 'float', 'Accuracy(%)': 'float', 'kmer': 'int'})

# Function to plot 3D scatter plots for accuracy and time comparison in the same figure
def plot_3d_scatter_comparison(data1, data2, x_col, y_col, z_col, title, ax):
    scatter1 = ax.scatter(data1[x_col], data1[y_col], data1[z_col], c='blue', label='Minimap2', alpha=0.6)
    scatter2 = ax.scatter(data2[x_col], data2[y_col], data2[z_col], c='red', label='VAT', alpha=0.6)
    ax.set_xlabel(x_col)
    ax.set_ylabel(y_col)
    ax.set_zlabel(z_col)
    ax.set_title(title)
    ax.legend()
    return scatter1, scatter2

fig = plt.figure(figsize=(15, 10))

# Plotting Marine Accuracy Comparison
ax1 = fig.add_subplot(221, projection='3d')
scatter1, scatter2 = plot_3d_scatter_comparison(
    combined_df[(combined_df['Tool'] == 'Minimap2') & (combined_df['Dataset'] == 'Marine')],
    combined_df[(combined_df['Tool'] == 'VAT') & (combined_df['Dataset'] == 'Marine')],
    'kmer', 'window_size', 'Accuracy(%)', 'Marine Dataset Accuracy Comparison', ax1
)

# Plotting Marine Time Comparison
ax2 = fig.add_subplot(222, projection='3d')
scatter3, scatter4 = plot_3d_scatter_comparison(
    combined_df[(combined_df['Tool'] == 'Minimap2') & (combined_df['Dataset'] == 'Marine')],
    combined_df[(combined_df['Tool'] == 'VAT') & (combined_df['Dataset'] == 'Marine')],
    'kmer', 'window_size', 'Time(s)', 'Marine Dataset Time Comparison', ax2
)

# Plotting Cami Accuracy Comparison
ax3 = fig.add_subplot(223, projection='3d')
scatter5, scatter6 = plot_3d_scatter_comparison(
    combined_df[(combined_df['Tool'] == 'Minimap2') & (combined_df['Dataset'] == 'Cami')],
    combined_df[(combined_df['Tool'] == 'VAT') & (combined_df['Dataset'] == 'Cami')],
    'kmer', 'window_size', 'Accuracy(%)', 'Cami Dataset Accuracy Comparison', ax3
)

# Plotting Cami Time Comparison
ax4 = fig.add_subplot(224, projection='3d')
scatter7, scatter8 = plot_3d_scatter_comparison(
    combined_df[(combined_df['Tool'] == 'Minimap2') & (combined_df['Dataset'] == 'Cami')],
    combined_df[(combined_df['Tool'] == 'VAT') & (combined_df['Dataset'] == 'Cami')],
    'kmer', 'window_size', 'Time(s)', 'Cami Dataset Time Comparison', ax4
)

plt.tight_layout()
plt.show()
```
